```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import cv2
import os
```

```python
# имена папок. они же - имена классов.
labels=['karabin','hryvnia','stepler','trash']
#извлекаем списки файлов для каждой папки.
filelists={}
for i in labels:
    filelists[i]=os.listdir(i)
```

```python
dataset=dict()
des_list=['orb','sift','brisk']
for i in des_list:
    dataset[i]=pd.read_csv(f"{i}_dataset.csv")
```

```python
X_train=dict()
y_train=dict()
X_test=dict()
y_test=dict()

from sklearn.model_selection import train_test_split
for i in des_list:
    temp=dataset[i].drop(columns=['pathname'])
    X_train[i], X_test[i], \
    y_train[i], y_test[i]=train_test_split(temp.drop(columns=['target']),temp['target'], \
                                           test_size=0.1, random_state=0,stratify=temp['target'])
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import *
pred=dict()
from joblib import dump, load

for i in des_list:
    clf=LogisticRegression(solver='newton-cg',C=0.01)
    clf.fit(X_train[i], y_train[i])
    from joblib import dump, load
    pred[i]=clf.predict(X_test[i])
    print(f"{i}:")
    print(clf.classes_)
    print(f"Common accuracy: {np.round( accuracy_score(y_test[i],pred[i]), 5)} "   )
    print(f"Precision:{np.round( precision_score(y_test[i],pred[i],average=None) , 5) }")
    print(f"Recall:{np.round( recall_score(y_test[i],pred[i],average=None), 5) }" )
    print(f"Confusion matrix:\n{confusion_matrix(y_test[i],pred[i])}")
    print("\n")
    
    dump(clf, f'written_models/{i}_logreg.joblib') 
```

```python
from sklearn.svm import SVC,LinearSVC
from sklearn.metrics import *
pred=dict()

for i in des_list:
    clf=SVC(C=100)
    clf.fit(X_train[i], y_train[i])
    pred[i]=clf.predict(X_test[i])
    print(f"{i}:")
    print(clf.classes_)
    print(f"Common accuracy: {np.round( accuracy_score(y_test[i],pred[i]), 5) } "   )
    print(f"Precision:{np.round( precision_score(y_test[i],pred[i],average=None), 5) }")
    print(f"Recall:{np.round(recall_score(y_test[i],pred[i],average=None),5) }" )
    print(f"Confusion matrix:\n{confusion_matrix(y_test[i],pred[i])}")
    print("\n")
    
    dump(clf, f'written_models/{i}_svm.joblib')
```

```python
#from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import *
pred=dict()

for i in des_list:
    clf=RandomForestClassifier(n_estimators=200)
    clf.fit(X_train[i], y_train[i])
    pred[i]=clf.predict(X_test[i])
    print(f"{i}:")
    print(clf.classes_)
    print(f"Common accuracy: {np.round( accuracy_score(y_test[i],pred[i]),5)} "   )
    print(f"Precision:{np.round(precision_score(y_test[i],pred[i],average=None),5)}")
    print(f"Recall:{np.round(recall_score(y_test[i],pred[i],average=None),5)}" )
    print(f"Confusion matrix:\n{confusion_matrix(y_test[i],pred[i])}")
    print("\n")
    
    dump(clf, f'written_models/{i}_randomforest.joblib')
```

```python
#from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import *
pred=dict()

for i in des_list:
    clf=KNeighborsClassifier(n_neighbors=5, weights='distance')
    clf.fit(X_train[i], y_train[i])
    pred[i]=clf.predict(X_test[i])
    print(f"{i}:")
    print(clf.classes_)
    print(f"Common accuracy: {np.round( accuracy_score(y_test[i],pred[i]),5)} "   )
    print(f"Precision:{np.round(precision_score(y_test[i],pred[i],average=None),5)}")
    print(f"Recall:{np.round(recall_score(y_test[i],pred[i],average=None),5)}" )
    print(f"Confusion matrix:\n{confusion_matrix(y_test[i],pred[i])}")
    print("\n")
    
    dump(clf, f'written_models/{i}_knn.joblib')
```
